# Average Sale Prices by Neighborhood

This script calculates the average sale price of houses in different neighborhoods using the dataset `df`.

```python

import matplotlib.pyplot as plt
avg_prices = df.groupby("Neighborhood")["SalePrice"].mean().reset_index()

```

FUTURE PREDICTIONS for better investment opportunities

```python
import folium
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# Define future years
future_years = np.array([2025, 2026, 2027, 2028, 2029]).reshape(-1, 1)

# Dictionary to store future prices per year
predicted_prices_per_year = {year: {} for year in future_years.flatten()}

# Train model and predict for each neighborhood
for neighborhood in avg_prices['Neighborhood']:
    temp_df = df[df["Neighborhood"] == neighborhood]

    if len(temp_df) > 1:  # Only train if there’s enough data
        model = LinearRegression()
        X = temp_df["YrSold"].values.reshape(-1, 1)
        y = temp_df["SalePrice"].values
        model.fit(X, y)

        # Predict for each future year
        future_prices = model.predict(future_years)

        for i, year in enumerate(future_years.flatten()):
            predicted_prices_per_year[year][neighborhood] = future_prices[i]

# Convert to DataFrame
future_prices_dfs = {
    year: pd.DataFrame(predicted_prices_per_year[year].items(), columns=["Neighborhood", "SalePrice"])
    for year in future_years.flatten()
}

# Merge each year with coordinates
merged_future_dfs = {
    year: future_prices_dfs[year].merge(coords_df, on="Neighborhood", how="left")
    for year in future_years.flatten()
}

# Create folium map
m = folium.Map(location=[42.034, -93.620], zoom_start=12)

# Base Layer for Current Prices (Blue)
current_layer = folium.FeatureGroup(name="Current Prices", show=True)
for _, row in merged_df.iterrows():
    folium.CircleMarker(
        location=[row["Latitude"], row["Longitude"]],
        radius=10,
        color="blue",
        fill=True,
        fill_color="blue",
        fill_opacity=0.6,
        popup=f"{row['Neighborhood']} (Current): ${row['SalePrice']:.2f}"
    ).add_to(current_layer)

m.add_child(current_layer)

# Future Year Layers (Different colors)
colors = ["red", "orange", "green", "purple", "brown"]
for i, year in enumerate(future_years.flatten()):
    future_layer = folium.FeatureGroup(name=f"Predicted Prices {year}", show=False)

    for _, row in merged_future_dfs[year].iterrows():
        folium.CircleMarker(
            location=[row["Latitude"], row["Longitude"]],
            radius=10,
            color=colors[i % len(colors)],
            fill=True,
            fill_color=colors[i % len(colors)],
            fill_opacity=0.6,
            popup=f"{row['Neighborhood']} ({year}): ${row['SalePrice']:.2f}"
        ).add_to(future_layer)

    m.add_child(future_layer)

# Add Layer Control
folium.LayerControl().add_to(m)

m
```
